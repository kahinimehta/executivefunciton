```python
#This script uploads BIDS-valid task data (.tsv's and .json's) to corresponding sessions in a flywheel project. 
#Inputs:
    #directory of converted task data (from PennBox)
#Outputs: 
    #none
```

```python
import glob
import flywheel
fw = flywheel.Client()
```

```python
#set project using name on flywheel 
EFProj=fw.projects.find_first('label=EFR01')
#print(EFProj)
```

```python
#iterate through subjects and sessions to access session containers 
subjects=EFProj.subjects()

sessions=[]
for s in subjects :
    tempsessions=s.sessions()
    sessions.extend(tempsessions)
```

```python
#isolate files without attachcments 
files=[]

for s in sessions:
    for f in range(len(s["files"])):
        file_name = (s["files"][f]["name"])
        if 'task-fracback_acq-singleband_bold_events.tsv' in file_name or 'task-fracback_acq-singleband_bold_events.json' in file_name:
            files.append(file_name)
```

```python
sessions_2=[]
for s in sessions:
    scanid=str(s.label)
    if not any(scanid in f for f in files):
        sessions_2.append(s)  
```

```python
#set directory to location of files for upload
root_dir = '/Users/krmurtha/Box Sync/EXECUTIVE_FUNCTION/SCANNER_TASK/' #change file path for appropriate user 
bids_dir = 'BIDS_2022'
all_files=glob.glob(root_dir + bids_dir + '*/*/*')
print(len(all_files))
```

```python
#iterate through sessions, find matching tsv, and upload! 
for ses in sessions_2:
    scanid=ses.label
    for file in all_files:
        #print(file)
        f=file.split('/')
        g=(f[len(f)-1]).split('_')
        h=(g[1].split('-'))[1]
        if scanid == h:
            print("uploading {0} to {1}".format(file, scanid))
            ses.upload_file(file)
```
